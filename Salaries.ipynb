```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'ds_salaries.csv'

# Read the CSV file with the correct parameters
dataset = pd.read_csv('ds_salaries.csv', encoding='ISO-8859-1', on_bad_lines='skip')

# Display the first few rows of the dataframe
dataset
```

```python
import pandas as pd


# Assuming 'job title' is the column name where job roles are listed
job_titles = dataset['job_title'].unique()

# Display the unique job titles
print("Job titles present in the dataset:")
for title in job_titles:
    print(title)

# Optionally, count occurrences of each job title
job_title_counts = dataset['job_title'].value_counts()
print("\nJob title frequencies:")
print(job_title_counts)
```

```python


# Group by 'job_role' and calculate average salary
average_salary = dataset.groupby('job_title')['salary'].mean().reset_index()

# Sort by average salary in descending order
sorted_by_salary = average_salary.sort_values(by='salary', ascending=False)

# Display job roles sorted by average salary from highest to lowest
print("Job roles sorted by average salary (from highest to lowest):")
for index, row in sorted_by_salary.iterrows():
    print(f"{row['job_title']}: ${row['salary']:.2f}")

```

```python

# Group by 'job_role' and count occurrences
employee_counts = dataset['job_title'].value_counts()

# Get the job role with the highest count of employees
highest_employee_job = employee_counts.idxmax()
highest_employee_count = employee_counts.max()

print(f"The job role with the highest number of employees is '{highest_employee_job}' with {highest_employee_count} employees.")
```

```python


# Assuming your dataset has columns like 'size_category', 'job_role'
# Group by 'size_category' and 'job_role', then count the number of employees
company_size_employee_counts = dataset.groupby(['company_size', 'job_title']).size().reset_index(name='employee_count')

# Find the size category with the most employees for each job role
max_employees_per_job_role = company_size_employee_counts.loc[company_size_employee_counts.groupby('job_title')['employee_count'].idxmax()]

# Display the size categories with the most employees for each job role
print("Size categories with the most employees for each job role:")
for index, row in max_employees_per_job_role.iterrows():
    print(f"Job role: {row['job_title']}, Size category: {row['company_size']}, Employees: {row['employee_count']}")



```

```python

# Assuming your dataset has columns like 'company_size' and 'job_title'
# Group by 'job_title' and find the mode (most frequent) company size
dominant_size_by_job_role = dataset.groupby('job_title')['company_size'].agg(lambda x: x.mode()[0]).reset_index()

# Display the dominant size category for each job role
print("Dominant size category for each job role:")
for index, row in dominant_size_by_job_role.iterrows():
    print(f"Job role: {row['job_title']}, Dominant size category: {row['company_size']}")
```

```python


# Find the mode (most frequent) company size across all job titles
dominant_company_size = dataset['company_size'].mode()[0]

print(f"The company size with the highest number of employees across job titles is '{dominant_company_size}'.")


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
datasets = pd.read_csv('ds_salaries.csv', encoding='ISO-8859-1', on_bad_lines='skip')
pd.set_option('display.max_rows',None)

# Display the first few rows of the data
print("First few rows of the dataset:")
print(datasets.head())

```

```python
# Summary of the dataset
print("\nSummary of the dataset:")
print(datasets.info())
print("\nDescriptive statistics:")
print(datasets.describe())
```

```python
# Initial data visualizations
plt.figure(figsize=(10, 6))
sns.countplot(data=datasets, x='experience_level')
plt.title('Distribution of Experience Levels')
plt.show()


```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=datasets, x='employment_type', y='salary_in_usd')
plt.title('Salary Distribution by Employment Type')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=datasets, x='job_title', y='salary_in_usd')
plt.title('Salary Distribution by Job Title')
plt.xticks(rotation=90)
plt.show()
```

```python
# Average Salary by Work Year and Experience Level
plt.figure(figsize=(12, 8))
sns.lineplot(data=datasets, x='work_year', y='salary_in_usd', hue='experience_level', marker='o')
plt.title('Average Salary by Work Year and Experience Level')
plt.xlabel('Work Year')
plt.ylabel('Average Salary in USD')
plt.legend(title='Experience Level')
plt.show()
```

```python
# Top Company Locations by Average Salary
top_company_locations = datasets.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_company_locations.values, y=top_company_locations.index)
plt.title('Top 10 Company Locations by Average Salary')
plt.xlabel('Average Salary in USD')
plt.ylabel('Company Location')
plt.show()
```

```python
# Distribution of Remote Work Ratios Across Job Titles
plt.figure(figsize=(14, 10))
sns.boxplot(data=datasets, x='remote_ratio', y='job_title')
plt.title('Distribution of Remote Work Ratios Across Job Titles')
plt.xlabel('Remote Work Ratio')
plt.ylabel('Job Title')
plt.xticks(rotation=90)
plt.show()
```
